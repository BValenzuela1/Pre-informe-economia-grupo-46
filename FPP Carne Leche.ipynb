```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, fixed

# initial parameters
ALPHA = 0.6   # capital share en producir carne
BETA = 0.2   # capital share en producir leche
KBAR = 1000  # Total capital supply
LBAR = 1000  # Total labor supply
p = 28.8      # =7000/243  precio relativo de a carne con respecto a la leche

def F(K,L,alpha=ALPHA):
    """Agriculture Production function"""
    return (K**alpha)*(L**(1-alpha))

def G(K,L,beta=BETA):
    """Manufacturing Production function"""
    return (K**beta)*(L**(1-beta))

def budgetc(c1, p1, p2, I):
    return (I/p2)-(p1/p2)*c1

def isoq(L, Q, mu):
    return (Q/(L**(1-mu)))**(1/mu)

def edgeworth(L, Kbar=KBAR, Lbar=LBAR,alpha=ALPHA, beta=BETA):
    """efficiency locus: """
    a = (1-alpha)/alpha
    b = (1-beta)/beta
    return b*L*Kbar/(a*(Lbar-L)+b*L)

def edgeplot(LA, Kbar=KBAR, Lbar=LBAR,alpha=ALPHA,beta=BETA):
    """Draw an edgeworth box
    
    arguments:
    LA -- labor allocated to ag, from which calculate QA(Ka(La),La) 
    """
    KA = edgeworth(LA, Kbar, Lbar,alpha, beta)
    RTS = (alpha/(1-alpha))*(KA/LA)
    QA = F(KA,LA,alpha)
    QM = G(Kbar-KA,Lbar-LA,beta)
    print("(LA,KA)=({:4.1f}, {:4.1f})  (QA, QM)=({:4.1f}, {:4.1f})  RTS={:4.1f}"
          .format(LA,KA,QA,QM,RTS))
    La = np.arange(1,Lbar)
    fig, ax = plt.subplots(figsize=(7,6))
    ax.set_xlim(0, Lbar)
    ax.set_ylim(0, Kbar)
    ax.plot(La, edgeworth(La,Kbar,Lbar,alpha,beta),'k-')
    #ax.plot(La, La,'k--')
    ax.plot(La, isoq(La, QA, alpha))
    ax.plot(La, Kbar-isoq(Lbar-La, QM, beta),'g-')
    ax.plot(LA, KA,'ob')
    ax.vlines(LA,0,KA, linestyles="dashed")
    ax.hlines(KA,0,LA, linestyles="dashed")
    ax.text(-6,-6,r'$O_A$',fontsize=16)
    ax.text(Lbar,Kbar,r'$O_M$',fontsize=16)
    ax.set_xlabel(r'$L_A -- Labor$', fontsize=16)
    ax.set_ylabel('$K_A - Capital$', fontsize=16)
    #plt.show()

def ppf(LA,Kbar=KBAR, Lbar=LBAR,alpha=ALPHA,beta=BETA):
    """Draw a production possibility frontier
    
    arguments:
    LA -- labor allocated to ag, from which calculate QA(Ka(La),La) 
    """
    KA = edgeworth(LA, Kbar, Lbar,alpha, beta)
    RTS = (alpha/(1-alpha))*(KA/LA)
    QA = F( KA,LA,alpha)
    QM = G(Kbar-KA,Lbar-LA,beta)
    ax.scatter(QA,QM)
    La = np.arange(0,Lbar)
    Ka = edgeworth(La, Kbar, Lbar,alpha, beta)
    Qa = F(Ka,La,alpha)
    Qm = G(Kbar-Ka,Lbar-La,beta)
    ax.set_xlim(0, Lbar)
    ax.set_ylim(0, Kbar)
    ax.plot(Qa, Qm,'k-')
    ax.set_xlabel(r'Producción carne',fontsize=18)
    ax.set_ylabel(r'Producción leche',fontsize=18)
    plt.show()

fig, ax = plt.subplots(figsize=(7,6))
ppf(400,alpha=0.6,beta=0.2)
```

**Recopilación de los costos más importantes**


**1. Alimentación**

Este es el mayor componente y el más importante en la crianza de ganado bovino. Gracias a una buena alimentación, el ganado es capaz de mantener una buena salud y un gran bienestar en general. Una dieta nutritiva y equilibrada es esencial para entregar al animal los nutrientes necesarios, por ejemplo proteína, carbohidratos, vitaminas y minerales, para un buen crecimiento y desarrollo. Un déficit alimenticio supone un mayor riesgo a enfermedades, un mal desarrollo y crecimiento, lo que conlleva a una disminución drástica en la calidad de la carne, resultando negativamente en la producción y rentabilidad del ganado. 


Además, una buena nutrición es esencial para una buena reproducción. Una buena dieta lleva a un buen desarrollo del feto y produce, en la vaca, leche de buena calidad para el crecimiento del mismo. Con esto se asegura una nueva generación más grande, más contenta y de mayor calidad. Con una mala alimentación, el ganado podría tener partos difíciles, el ternero podría nacer con enfermedades, habría una menor producción de leche y una disminución de la fertilidad en el ganado. 


En resumen, una alimentación equilibrada y de calidad significa una buena reproducción, crecimiento y desarrollo del ganado. Mejora en la calidad de la carne y en los nutrientes que ésta aporta. Una buena alimentación es clave para el éxito, la eficiencia y la rentabilidad de la crianza bovina.


**2. Manejo sanitario**


Al igual que la alimentación, el manejo sanitario es muy importante para mantener la salud y el bienestar del ganado. Incluyendo la prevención y control de enfermedades, la vacunación y el seguimiento de la salud de la crianza. Un ganado sano tiene mejor crecimiento, desarrollo y reproducción, lo que, como ya lo vimos anteriormente, es clave para una buena eficiencia y rentabilidad.


Un mal manejo sanitario conlleva a grandes pérdidas económicas como por ejemplo la muerte de un animal, costosos tratamientos, disminución en la producción de carne o leche, entre otros. Un buen manejo sanitario ayuda de gran manera a prevenir, o minimizar, estos costos indeseados. 


Dependiendo de la región en la que se críe el ganado, se debe tener en cuenta las normativas y regulaciones específicas en esa región sobre la crianza de ganado bovino. Es clave un buen manejo sanitario para poder operar legalmente en el territorio y evitar multas y sanciones, las cuales, además de ser un gran gasto económico, hacen insostenible e inviable el negocio a largo plazo.


**3. Infraestructura y mano de obra**


Una infraestructura adecuada, como por ejemplo corrales y cercas bien diseñadas, generan un ambiente seguro y cómodo para los animales. Además, permite que los animales se manejen con más eficiencia y seguridad, reduciendo así el estrés y el riesgo de lesiones. Una mano de obra bien capacitada se asegura de que los animales reciban atención médica adecuada y de calidad, y sean alimentados de manera adecuada de acuerdo a sus necesidades. Sumando a la seguridad y comodidad del ganado, una buena infraestructura y mano de obra pueden mejorar la eficiencia en la producción de carne y leche. 


Los corrales bien diseñados permiten separar el ganado en categorias de edad y sexo, mejorando la tasa de crecimiento porque se reduce la competencia por el alimento y agua, además de recibir una nutrición adecuada según su edad. Las instalaciones de almacenamiento y procesamiento de alimentos bien estructuradas previenen la contaminación y la propagación de enfermedades. La mano de obra se encarga de la mantención de estas máquinas y de seguir los protocolos de seguridad e higiene en todas las etapas de producción de carne.

**Estimación elasticidad de oferta y demanda en la crianza de ganado bovino**


**1. Elasticidad de oferta**


La elasticidad de oferta es un concepto que se utiliza en economía para referirse a la capacidad de los productores de ganado para ajustar la cantidad de ganado que ofrecen en el mercado como respuesta a variaciones en el precio del ganado. La elasticidad se puede ver afectada por diversos factores, tales como el costo de insumos, las fluctuaciones en el precio del ganado y la oferta de ganado importado. Si los costos de producción son bajos y los precios del ganado son altos, habrá una mayor motivación para aumentar la oferta de ganado en el mercado. En el caso contrario, donde los costos son altos y los precios bajos, no habrá motivación para producir carne, por lo que la oferta de ganado disminuye.


Una elasticidad de oferta alta supone que los productores pueden ajustar rápidamente la cantidad de ganado que ofrecen en el mercado. Por otro lado, una elasticidad baja indica que los productores tienen dificultades para ajustar la producción en respuesta a los cambios en el precio, lo que resulta en una oferta más rígida y menos sensible a los cambios en el mercado. 


En Chile, la elasticidad de oferta de la cría bovina en Chile se ha mantenido estable en los últimos años, pero hay que destacar que la elasticidad puede variar según la región, por lo que es necesario considerar estos factores al analizar el comportamiento del mercado. En general, la elasticidad de oferta de la cría bovina en Chile es relativamente baja debido a los altos costos de producción y la limitada disponibilidad de insumos. En 2021, la producción nacional de carne bovina fue de 209.941 toneladas, mientras que en 2022, fue de 196.924 toneladas, es decir, tuvo un uan caída en la producción de un 
6,3%. Por otro lado, la variación del precio entre 2021 y 2022 fue de un 27%, pudiendo así calcular que la elasticidad precio de la oferta es 0,23. Por lo que se puede concluir que la oferta es de la cría bovina es inelástica.



**2. Elasticidad de demanda**

La elasticidad de la demanda es un término económico que se refiere a la medida en que la cantidad demandada de carne de vacuno varía en respuesta a cambios en el precio de mercado, es decir, es la sensibilidad de la cantidad demandada a los cambios en el precio. Esta puede verse afectada por diversos factores, tales como el nivel de ingresos de la población, las preferencias de los consumidores, la disponibilidad de sustitutos, entre otros. Si los precios de la carne de vacuno aumentan, los consumidores pueden optar por comprar otros productos sustitutos, como pollo o cerdo, lo que reduce la cantidad demandada de carne de vacuno.


La elasticidad de la demanda en la cría bovina en Chile se ha mantenido estable en los últimos años, pero se han observado fluctuaciones en respuesta a cambios en la oferta y en los precios. La elasticidad de la demanda de carne de vacuno en Chile es moderada a baja, lo que significa que la cantidad demandada no se ve afectada en gran medida por los cambios en el precio.
```python
class mercado:

    def __init__(self, DAPmax, COmin, d, s):
        """
        Configura los parámetros del mercado. DAP𝑚𝑎𝑥 es la máxima disposición a 
        pagar de cualquier consumider-el intercepto con el eje y de la función 
        de demanda. COmin es el mínimo costo oportunidad de cualquier productor-el 
        intercepto con el eje y de la función de oferta. s es la pendiente de la f
        unción de oferta. d es la pendiente de la función de demanda...
        """
        
        self.DAPmax, self.COmin, self.d, self.s = DAPmax, COmin, d, s
        
        if DAPmax < COmin:
            raise ValueError('Demanda insuficiente.')

    def cantidad_equilibrio(self):
        "Cálculo cantidad de equilibrio"
        return  (self.DAPmax - self.COmin)/(self.d + self.s)

    def precio_equilibrio(self):
        "Cálculo cantidad de equilibrio"
        return  self.DAPmax - self.d * self.cantidad_equilibrio()

    def excedente_consumidor(self):
        "Cálculo excedente consumidor"
        return (self.DAPmax - self.precio_equilibrio())*self.cantidad_equilibrio()/2

    def excedente_productor(self):
        "Cálculo excedente productor"
        return (self.precio_equilibrio() - self.COmin) * self.cantidad_equilibrio() /2

    def excedente_total(self):
        "Cálculo excedente total"
        return self.excedente_productor() + self.excedente_consumidor()
    
    def demanda(self,x):
        "Función demanda"
        return self.DAPmax - self.d*x
        
    def oferta(self,x):
        "Función oferta"
        return self.COmin + self.s*x

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Parámetros iniciales DAPmax, COmin, d, s
params_ini = 10000,1,(515/64020),(515/13017)
m = mercado(*params_ini)

q_max = m.cantidad_equilibrio() * 2
q_grid = np.linspace(0.0, q_max, 100)
pd = m.demanda(q_grid)
ps = m.oferta(q_grid)

fig, ax = plt.subplots()
ax.plot(q_grid, pd, lw=2, alpha=0.6, label='demanda')
ax.plot(q_grid, ps, lw=2, alpha=0.6, label='oferta')
ax.set_xlabel('Cantidad', fontsize=14)
ax.set_xlim(0, q_max)
ax.set_ylabel('Precio', fontsize=14)
ax.legend(loc='lower right', frameon=False, fontsize=14)
ax.set(title='Oferta, Demanda y Equilibrio de mercados')
plt.show()

# Baseline DAPmax, COmin, d, s
print("Excedente productor =",m.excedente_productor())
print("Excedente consumidor =",m.excedente_consumidor())
print("Excedente total =",m.excedente_total())
print("Cantidad de equilibrio =",m.cantidad_equilibrio())
print("Precio de equilibrio =",m.precio_equilibrio())
```
